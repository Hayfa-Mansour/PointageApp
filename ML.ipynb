```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# 1. Chargement et Préparation "Réelle" des données
df = pd.read_csv('Depression-Dataset.csv')

# Encodage de toutes les colonnes textuelles (Facteurs réels)
le = LabelEncoder()
for col in df.columns:
    if df[col].dtype == 'object':
        df[col] = le.fit_transform(df[col])

# Sélection de toutes les colonnes de facteurs pour la prédiction
X = df.drop('DEPRESSED', axis=1)
y = df['DEPRESSED']

# Mise à l'échelle (Standardisation) - Crucial pour KNN et Logistic Regression
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Division Train/Test
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# 2. Définition des 4 modèles
models = {
    "Logistic Regression": LogisticRegression(),
    "Decision Tree": DecisionTreeClassifier(max_depth=5),
    "Random Forest": RandomForestClassifier(n_estimators=100),
    "K-Nearest Neighbors": KNeighborsClassifier(n_neighbors=5)
}

# Stockage des résultats
results = {}
conf_matrices = {}

for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    results[name] = accuracy_score(y_test, y_pred)
    conf_matrices[name] = confusion_matrix(y_test, y_pred)

# 3. PRÉSENTATION VISUELLE
plt.figure(figsize=(18, 12))

# --- A. Comparaison des scores d'Exactitude (Accuracy) ---
plt.subplot(2, 2, 1)
sns.barplot(x=list(results.keys()), y=list(results.values()), palette='coolwarm')
plt.title('Comparaison de la Précision (Accuracy)', fontsize=14)
plt.ylim(0, 1)
for i, v in enumerate(results.values()):
    plt.text(i, v + 0.02, f"{v:.2%}", ha='center', fontweight='bold')

# --- B. Importance des Facteurs (via Random Forest) ---
# Pour montrer quelles colonnes sont les plus "réelles" dans la prédiction
plt.subplot(2, 2, 2)
rf = RandomForestClassifier().fit(X, y) # On utilise X non-scalé pour les noms de colonnes
importances = pd.Series(rf.feature_importances_, index=X.columns).sort_values(ascending=False).head(10)
sns.barplot(x=importances.values, y=importances.index, palette='viridis')
plt.title('Top 10 des Facteurs Déterminants', fontsize=14)

# --- C. Matrices de Confusion (Focus sur les 4 modèles) ---
# On affiche les matrices pour voir où les modèles se trompent
for i, (name, matrix) in enumerate(conf_matrices.items()):
    plt.subplot(2, 4, i + 5) # Utilise la ligne du bas
    sns.heatmap(matrix, annot=True, fmt='d', cmap='Blues', cbar=False)
    plt.title(f'Matrix: {name}', fontsize=10)
    plt.xlabel('Prédiction')
    plt.ylabel('Réalité')

plt.tight_layout()
plt.show()

# 4. Affichage du rapport détaillé dans la console
for name, model in models.items():
    print(f"\n--- Rapport pour {name} ---")
    print(classification_report(y_test, model.predict(X_test)))
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler

# Chargement
df = pd.read_csv('Depression-Dataset.csv')

# Encodage de toutes les colonnes pour que la prédiction soit basée sur des faits réels
le = LabelEncoder()
for col in df.columns:
    if df[col].dtype == 'object':
        df[col] = le.fit_transform(df[col])

X = df.drop('DEPRESSED', axis=1)
y = df['DEPRESSED']

# Normalisation (Important pour certains modèles)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X) #Le StandardScaler calcule la moyenne et l'écart-type de toutes les données contenues dans X pour normaliser.

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
scores = {} # Pour stocker les résultats pour la comparaison finale
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

model_lr = LogisticRegression()
model_lr.fit(X_train, y_train)
y_pred_lr = model_lr.predict(X_test)
scores['Régression Logistique'] = accuracy_score(y_test, y_pred_lr)

# Visualisation : Matrice de Confusion
ConfusionMatrixDisplay.from_estimator(model_lr, X_test, y_test, cmap='Reds')
plt.title("Matrice de Confusion : Régression Logistique")
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree

model_dt = DecisionTreeClassifier(max_depth=3) # Profondeur limitée pour la visibilité
model_dt.fit(X_train, y_train)
y_pred_dt = model_dt.predict(X_test)
scores['Arbre de Décision'] = accuracy_score(y_test, y_pred_dt)

# Visualisation : L'arbre lui-même
plt.figure(figsize=(15,8))
plot_tree(model_dt, feature_names=X.columns, class_names=['Sain', 'Déprimé'], filled=True)
plt.title("Visualisation des règles de décision")
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
import seaborn as sns

model_rf = RandomForestClassifier(n_estimators=100)
model_rf.fit(X_train, y_train)
y_pred_rf = model_rf.predict(X_test)
scores['Random Forest'] = accuracy_score(y_test, y_pred_rf)

# Visualisation : Importance des variables réelles
importances = pd.Series(model_rf.feature_importances_, index=X.columns).sort_values(ascending=False)
plt.figure(figsize=(10,6))
sns.barplot(x=importances.head(10), y=importances.head(10).index)
plt.title("Top 10 des facteurs réels causant la dépression")
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier

model_knn = KNeighborsClassifier(n_neighbors=5)
model_knn.fit(X_train, y_train)
y_pred_knn = model_knn.predict(X_test)
scores['KNN'] = accuracy_score(y_test, y_pred_knn)

# Visualisation : Matrice de Confusion
ConfusionMatrixDisplay.from_estimator(model_knn, X_test, y_test, cmap='Greens')
plt.title("Matrice de Confusion : KNN")
plt.show()
```

```python
import seaborn as sns

# Création du graphique comparatif
plt.figure(figsize=(10, 6))
ax = sns.barplot(x=list(scores.keys()), y=list(scores.values()), palette='magma')
plt.title('COMPARAISON FINALE DES MODÈLES', fontsize=16)
plt.ylabel('Précision (Accuracy)')
plt.ylim(0, 1)

# Ajouter les pourcentages sur les barres
for p in ax.patches:
    ax.annotate(f'{p.get_height():.2%}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points', fontweight='bold')

plt.show()

# Conclusion textuelle
best_m = max(scores, key=scores.get)
print(f"Le meilleur modèle pour cette problématique est : {best_m} avec {scores[best_m]:.2%} de précision.")
```
